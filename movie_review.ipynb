```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
unique_sent = df['sentiment'].unique()
sent_num = {}
i = 0
for sn in unique_sent:
    sent_num[sn]=i
    i+=1
df['sentiment'] = df['sentiment'].map(sent_num)
```

```python
df
```

```python
df['review'] = df['review'].apply(lambda x: x.lower())
```

```python
import string
def remove_punc(txt):
    return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['review'] = df['review'].apply(remove_punc)
```

```python
def remove_num(txt):
    st = ""
    for i in txt:
        if not i.isdigit():
            st+=i
    return st 
```

```python
df['review'] = df['review'].apply(remove_num)
```

```python
def remove_emoji(txt):
    st=""
    for i in txt:
        if i.isascii():
            st+=i
    return st 
```
