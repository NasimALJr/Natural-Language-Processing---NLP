```python
# lowercase a set of text columns safely (handle NaNs by converting to string first)
cols = ['course_code','course_name','material_type','material_title','description','keywords']
df[cols] = df[cols].astype(str).apply(lambda col: col.str.lower())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df['description'], df['material_link'], test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Create a bag-of-words vectorizer and fit on training text
bow = CountVectorizer()
x_train_bow = bow.fit_transform(X_train)

# Train the classifier
nb_model = MultinomialNB()
nb_model.fit(x_train_bow, y_train)

# Example prediction: transform text before predicting
sample = ["Operating System Book"]
sample_vec = bow.transform(sample)
print('Sample prediction:', nb_model.predict(sample_vec))

# Evaluate on the test set - transform rather than re-fitting the vectorizer
x_test_bow = bow.transform(X_test)
test_preds = nb_model.predict(x_test_bow)
print('Accuracy:', accuracy_score(y_test, test_preds))

```
