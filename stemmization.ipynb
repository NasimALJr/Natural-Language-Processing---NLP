```python
from nltk.stem import PorterStemmer
stm = PorterStemmer()
words = ['Avoided','Believable','Taking','Training']
for i in words:
    print(i,'-->',stm.stem(i))
    
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp("I am working since morning, want to work next monring bathing ate eaten avoided")
for i in doc:
    print(i,'-->',i.lemma_)
```

```python
nlp.pipe_names
```

```python
at = nlp.get_pipe('attribute_ruler')
at.add([[{'TEXT':'Bro'}],[{'TEXT':'Bra'}]],{'LEMMA':'Brother'})
```

```python
doc = nlp("I am working since morning Bro, Bra want to work next monring bathing ate eaten avoided Brother")
for i in doc:
    print(i,'-->',i.lemma_)
```
