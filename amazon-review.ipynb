```python
import bz2

train_path = "/kaggle/input/amazonreviews/train.ft.txt.bz2"

with bz2.open(train_path, "rt", encoding="utf-8") as f:
    lines = f.readlines()

print(lines[2])
```

# Converting Data to a DataFrame

```python
import pandas as pd

texts = []
labels = []

for line in lines:
    line = line.strip()
    if not line:
        continue

    label, text = line.split(" ", 1)
    labels.append(label)      
    texts.append(text)

df = pd.DataFrame({
    "text": texts,
    "label": labels
})

df.head()
```

# Labeling 'label' column 1--> negative && 0--> positive

```python
lb = df['label'].unique()
lb_num = {}
j=0
for i in lb:
    lb_num[i] = j;
    j=j+1
df['label'] = df['label'].map(lb_num)
df.sample(5)
```

# converting all text to lower 

```python
df['text'] = df['text'].apply(lambda x: x.lower())
df.head(10)
```

# Removing punctuation marks

```python
import string

def remove_punc(txt):
    st=""
    for ch in txt:
        if ch not in string.punctuation:
            st=st+ch
    return st
df['text'] = df['text'].apply(remove_punc)
df.head(10)
    
```

# Removing Number from 'text' column

```python
def remove_num(txt):
    st=""
    for ch in txt:
        if not ch.isdigit():
            st=st+ch
    return st
df['text'] = df['text'].apply(remove_num)
df.head(10)
        
```

# Removing emoji's

```python
def remove_emoji(txt):
    st=""
    for ch in txt:
        if ch.isascii():
            st=st+ch
    return st
df['text'] = df['text'].apply(remove_emoji)
df.head(10)
```

# Removing extra space

```python
df['text'] = df['text'].str.replace(r"\s+", " ", regex=True).str.strip()
df.head(10)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download("stopwords")
nltk.download("punkt")
```

# Removing Stopwords

```python
stop_words = set(stopwords.words('english'))

def remove_sw(txt):
    words = txt.split()
    cleaned = []
    for i in words:
        if i.lower() not in stop_words:
            cleaned.append(i)
    return ' '.join(cleaned)

df['text'] = df['text'].apply(remove_sw)
```

```python
df.head(10)
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['text'],df['label'],test_size=0.2,random_state=42)
```

# Applying classifications Algorithms

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,f1_score
```

# Tokenization

```python
tf_idf = TfidfVectorizer()
X_train_tf = tf_idf.fit_transform(X_train)
X_test_tf = tf_idf.transform(X_test) 
```

# Applying Naive-Bayes

```python
nasim = MultinomialNB()
nasim.fit(X_train_tf,y_train)
nasim_pr = nasim.predict(X_test_tf)
```

```python

print(f"Accuracy using Naive Bayes : {accuracy_score(y_test,nasim_pr)}")
```

# Applying LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_tf,y_train)
pred_lr = lr.predict(X_test_tf)
```

```python
print(f"Accuracy using Logistic Regression: {accuracy_score(y_test,pred_lr)}")
```
